```python
# Creating a random tensor with values between 0 & 100 with 50 values
myrand = tf.constant(np.random.randint(0,100, size = 50))
myrand
```

```python
tf.size(myrand), myrand.ndim, myrand.dtype, myrand.shape 
```

```python
# Minimum value
tf.reduce_min(myrand)
```

```python
# Maximum Value
tf.reduce_max(myrand)
```

```python
# Mean Value
tf.reduce_mean(myrand)
```

```python
# Sum
tf.reduce_sum(myrand)
```

```python
import tensorflow_probability as tfp
```

```python
# variance
tfp.stats.variance(myrand)
```

```python
tf.math.reduce_variance(tf.cast(myrand, dtype = tf.float32))
```

```python
# Standard Deviaiton
tf.math.reduce_std(tf.cast(myrand, dtype = tf.float32))
```
