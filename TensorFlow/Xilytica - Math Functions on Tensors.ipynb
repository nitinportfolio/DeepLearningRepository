# Squaring, Log, Square Root

```python
mytf = tf.range(1,10)
mytf
```

```python
tf.square(mytf)
```

```python
# tf.sqrt(mytf)
# Method has to be float
```

```python
# Method requires float dtypes
tf.sqrt(tf.cast(mytf, dtype = tf.float32))
```

```python
# Log
tf.math.log(tf.cast(mytf, dtype = tf.float32))
```
