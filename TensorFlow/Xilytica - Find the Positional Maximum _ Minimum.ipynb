```python
# Importing Tensorflow
import tensorflow as tf
print(tf.__version__)
import numpy as np
```

# Finding the positional maximum & minimum

```python
# Create a new tensor for finding positional minimum & maximum
tf.random.set_seed(101)
myrand = tf.random.uniform(shape = [50])
myrand
```

```python
# Find the positional maximum
tf.argmax(myrand)
```

```python
# Index on our largest value position
myrand[34]
```

```python
tf.math.reduce_max(myrand)
```

```python
myrand[tf.argmax(myrand)] == tf.math.reduce_max(myrand)
```

```python
assert myrand[tf.argmax(myrand)] == tf.math.reduce_max(myrand)
```

```python
# Finding the positional minimum
tf.argmin(myrand)
```

```python
myrand[tf.argmin(myrand)]
```

```python
tf.math.reduce_min(myrand)
```
