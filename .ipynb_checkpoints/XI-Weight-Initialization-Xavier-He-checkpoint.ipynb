```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/ushape.csv")
df.head()
```

```python
plt.scatter(df['X'],df['Y'],c=df['Class'])
```

```python
X = df.iloc[:,:2].values
y = df.iloc[:,-1].values
```

```python
plt.scatter( x = X[:,0],y=X[:,1], c=y)
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(10, activation="tanh", input_dim = 2))
model.add(Dense(10,activation="tanh"))
model.add(Dense(10,activation="tanh"))
model.add(Dense(10,activation="tanh"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
# Weights are initialised & bias are kept to zeros
model.get_weights()
```

```python
plt.plot(model.get_weights()[2],".")
plt.show()
```

# Without initializing weights

```python
model.compile(loss = "binary_crossentropy", optimizer = "adam",metrics=["accuracy"])
```

```python
history = model.fit(X,y,epochs = 100, validation_split=0.2)
```

```python
from sklearn.metrics import accuracy_score

y_prob = model.predict(X)
```

```python
y_pred = np.where(y_prob>=0.5,1,0)
```

```python
accuracy_score(y, y_pred)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X,y.astype('int'), clf=model, legend=2)
```

# With Initialization - 
## Xavier - Normalised Weights

```python
initial_weights = model.get_weights()
```

```python
# Changing weights for initialization
initial_weights[0] = np.random.randn(2,10)*np.sqrt(1/2)
initial_weights[1] = np.zeros(model.get_weights()[1].shape)
initial_weights[2] = np.random.randn(10,10)*np.sqrt(1/10)
initial_weights[3] = np.zeros(model.get_weights()[3].shape)
initial_weights[4] = np.random.randn(10,10)*np.sqrt(1/10)
initial_weights[5] = np.zeros(model.get_weights()[5].shape)
initial_weights[6] = np.random.randn(10,10)*np.sqrt(1/10)
initial_weights[7] = np.zeros(model.get_weights()[7].shape)
initial_weights[8] = np.random.randn(10,1)*np.sqrt(1/10)
initial_weights[9] = np.zeros(model.get_weights()[9].shape)
```

```python
initial_weights[0]
```

```python
np.random.randn(2,10)*np.sqrt(1/2)
```

```python
model.get_weights()[1].shape
```

```python
model.set_weights(initial_weights)
```

```python
model.get_weights()
```

```python
plt.plot(model.get_weights()[2],".")
plt.show()
```

```python
model.compile(loss = "binary_crossentropy", optimizer = "adam",metrics=["accuracy"])
```

```python
history = model.fit(X,y,epochs = 100, validation_split=0.2)
```

```python
y_pred = np.where(y_prob>=0.5,1,0)
```

```python
accuracy_score(y, y_pred)
```

# TensorFlow - Weights Initializations

### Based on how you want to initailize you can change 

**kernel_initalization**
 - he_normal/
 - he_uniform/
 - glorat_normal/
 - glorat_uniform
 
 **Glorat Uniform is the default weight initialization method**

```python
model = Sequential()
model.add(Dense(10, activation="relu", input_dim = 2, kernel_initializer="he_normal"))
model.add(Dense(10,activation="relu",kernel_initializer="he_normal"))
model.add(Dense(10,activation="relu",kernel_initializer="he_normal"))
model.add(Dense(10,activation="relu",kernel_initializer="he_normal"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(loss = "binary_crossentropy", optimizer = "adam",metrics=["accuracy"])
```

```python
history = model.fit(X,y,epochs = 100, validation_split=0.2)
```

```python
y_prob = model.predict(X)
```

```python
y_pred = np.where(y_prob>=0.5,1,0)
```

```python
accuracy_score(y, y_pred)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X,y.astype('int'), clf=model, legend=2)
```
